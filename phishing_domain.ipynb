```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowprice,uppprice = remove_outlier(train['qty_percent_url'])
train['qty_percent_url']=np.where(train['qty_percent_url']>uppprice,uppprice,train['qty_percent_url'])
train['qty_percent_url']=np.where(train['qty_percent_url']<lowprice,lowprice,train['qty_percent_url'])
```
